```python
# python countdown function

import time

my_time = int(input("Enter time in seconds: "))

for x in range(my_time, 0, -1):
    seconds = x % 60
    minutes = int(x / 60) % 60
    hours = int(x / 3600) % 24
    print(f"{hours:02}:{minutes:02}:{seconds:02}")
    time.sleep(1)

print("times up!")

```

```python
# python quiz game

questions = ("How many elements in the periodic table?:",
              "Which animal lays the largest eggs?: ",
              "What is the most abundant gas in Earths atmosphere?: ",
              "How many bones are in the human body?: ",
              "Which planet in the solar system is hottest?: ",)

options = (("A. 116", "B. 117", "C. 118", "D. 119"),
          ("A. platypus", "B. duck", "C. ostrich", "D. snake"),
          ("A. Nitrogen", "B. Oxygen", "C. Argon", "D. Hydrogen"),
          ("A. 206", "B. 207", "C. 208", "D. 209"),
          ("A. Venus", "B. Mars", "C. Mercury", "D. Jupiter"))

answers = ("C", "C", "D", "A", "A",)
guesses = []
score = 0
question_num = 0

for question in questions:
    print("------------------------")
    print(question)
    for option in options[question_num]:
        print(option)


    guess = input("Enter (A, B, C, D): ").upper()
    guesses.append(guess)
    if guess == answers[question_num]:
        score += 1
        print("CORRECT!")
    else:
        print("INCORRECT!")
        print(f"{answers[question_num]} is the correct answer")
    question_num += 1
```

```python
# Concession Stand Program

menu = {"pizza": 3.00,
        "nachos": 4.40,
        "popcorn": 6.00,
        "fries": 2.50,
        "chips": 1.00,
        "pretzel": 3.50,
        "soda": 3.00,
        "lemonade": 4.25}

cart = []
total = 0
print("--------MENU------------")
for key, value in menu.items():
    print(f"{key:10}: ${value:.2f}")
print("------------------------")

while True:
    food = input("Select an item (q to quit): ").lower()
    if food == "q":
        break
    elif menu.get(food) is not None:
        cart.append(food)

print("--------YOUR ORDER------------")
for food in cart:
    total += menu.get(food)
    print(food, end=" ")

print()
print(f"Total is: ${total:.2f}")

```

```python
# rock paper scissor

import random

options = ("rock", "paper", "scissors")
running = True

while running:

  player = None
  computer = random.choice(options)

  while player not in options:
    player = input("Enter a choice (rock, paper, scissors): ")

  print(f"Player: {player}")
  print(f"Computer: {computer}")

  if player == computer:
    print("It's a tie!")
  elif player == "rock" and computer == "scissors":
    print("You Win!")
  elif player == "paper" and computer == "rock":
    print("You Win!")
  elif player == "scissors" and computer == "paper":
    print("You win!")
  else:
    print("You lose!")

    if not input("Play again? (y/n): ").lower() == "y":
      running = False


print("Thanks for playing!")

```
